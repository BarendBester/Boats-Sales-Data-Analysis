### 6. Create a categorical plot and interpret the results.

```python
# Highest boat price by engine type

# 1. set figure size
fig = plt.figure(figsize=(17,10) )

# 2. group by the type of boat and the mean value in euro
engine_price = df_boats.groupby('type')['value_in_eur'].mean().sort_values().reset_index(drop = False)

# 3. Plot the graph
sns.barplot(x ='type', y = 'value_in_eur', data = engine_price,
                 palette = 'crest')
plt.title("Average Boat Price")
plt.xticks(rotation = 75)
```

```python
# Highest viewed by engine type

# 1. set figure size
fig = plt.figure(figsize=(17,10) )

# 2. group by the type of boat and the mean value in euro
engine_viewed = df_boats.groupby('type')['views_last_7_days'].mean().sort_values().reset_index(drop = False)

# 3. Plot the graph
engine = sns.barplot(x ='type', y = 'views_last_7_days', data = engine_viewed,
                 palette = 'crest')
plt.title("Average Boat - popularity by engine-type")
plt.xticks(rotation = 75)

engine.figure.savefig(os.path.join(path, '03-Analysis','Visualizations', 'engine&views.png'))
```

```python
df_boats.columns
```

```python
# Highest viewed by Material

# 1. set figure size
fig = plt.figure(figsize=(17,10) )

# 2. group by the type of boat and the mean value in euro
mat_viewed = df_boats.groupby('material')['views_last_7_days'].mean().sort_values().reset_index(drop = False)

# 3. Plot the graph
sns.barplot(x ='material', y = 'views_last_7_days', data = mat_viewed,
                 palette = 'crest')
plt.title("Average Boat - popularity by material")
plt.xticks(rotation = 75)

# plastic boats most popular
```

```python
# Highest price by Manufacturer

# 1. set figure size
fig = plt.figure(figsize=(17,10) )

# 2. group by the type of boat and the mean value in euro
mat_viewed = df_boats.groupby('material')['value_in_eur'].mean().sort_values().reset_index(drop = False)

# 3. Plot the graph
sns.barplot(y ='material', x = 'value_in_eur', data = mat_viewed,
                 palette = 'crest')
plt.title("Average Boat Price By Material")


# plastic boats most popular
```

```python
#see the quatiles for price
#create new variables based on these prices
df_boats['value_in_eur'].quantile([0.25, 0.50, 0.75])
```

```python
# create a new variable for the price of boats
# 1. low priced 
df_boats.loc[df_boats['value_in_eur'] <  44993.75, 'Price category'] = 'Low price'
```

```python
# 2. Medium priced 
df_boats.loc[(df_boats['value_in_eur'] >= 44993.75) & 
             (df_boats['value_in_eur'] < 95829.50), 'Price category'] = 'Middle price'
```

```python
# 3. high priced 
df_boats.loc[df_boats['value_in_eur'] > 95829.50, 'Price category'] = 'High price'
```

```python
df_boats['Price category'].value_counts()
```

```python
fig = plt.figure(figsize=(10,5) )

sns.barplot(y = 'Price category', x = 'views_last_7_days', data = df_boats,
                 palette = 'crest')

plt.title('Average views last 7 days per price category')

```

```python
# see quartiles for number of views
# create a new variable of number of views
df_boats['views_last_7_days'].quantile([0.25, 0.50, 0.75])
```

```python
# low popularity
df_boats.loc[df_boats['views_last_7_days'] <  70, 'Popularity'] = 'Not that popular'
```

```python
# medium popularity
df_boats.loc[(df_boats['views_last_7_days'] >= 70) & 
             (df_boats['views_last_7_days'] < 107), 'Popularity'] = 'More popular'
```

```python
# highly popular
df_boats.loc[df_boats['views_last_7_days'] > 170, 'Popularity'] = 'Very popular'
```

```python
df_boats['Popularity'].value_counts()
```

```python
fig = plt.figure(figsize=(10,5) )

sns.barplot(x = 'value_in_eur', y = 'Popularity', data = df_boats,
                 palette = 'crest')

plt.title('Popularity & Price')
```
